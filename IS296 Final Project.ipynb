```python
vote_percentage_array = health_econ_by_county_cleaned.column('Vote Percentage')
premature_death_array = health_econ_by_county_cleaned.column('premature_deathDeaths')

correl_vote_death = Table().with_column('Voting Percentage', vote_percentage_array, 'Premature Deaths', premature_death_array)
```

```python
correl_vote_death 
```

```python
clean_correl_vote_death  = correl_vote_death.where('Voting Percentage', are.not_equal_to('None Applicable')).where('Premature Deaths', are.not_equal_to('None Applicable'))
#I had to isolate the table to two columns to get the correlation coefficient due to the fact that any function would screw up due to null values
#It was also necessary due to the fact that removing all null values from the table would cause over 1000 peices of data to be lost
```

```python
def correlation_coefficient(table, col1, col2):
    col1_values = table.column(col1).astype(float)
    col2_values = table.column(col2).astype(float)

    return np.corrcoef(col1_values, col2_values)[0, 1]


correlation_coefficient(clean_correl_vote_death, 'Voting Percentage', 'Premature Deaths')
```

```python
def quartile(table, column):
    converted_column = table.column(column).astype(float)
    
    clean_table = table.with_column(column, converted_column)
    
    q1 = percentile(25, converted_column)
    q3 = percentile(75, converted_column)
    iqr = q3 - q1
    threshold = iqr * 1.5
    upper_threshold = q3 + threshold
    lower_threshold = q1 - threshold
    
    outliers_upper = clean_table.where(column, are.above(upper_threshold))
    outliers_lower = clean_table.where(column, are.below(lower_threshold))
    num_of_outliers = outliers_upper.num_rows + outliers_lower.num_rows
    
    return num_of_outliers, q1, q3
```
